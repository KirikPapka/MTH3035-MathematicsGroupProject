```python
# Atmospheric Profile Plotter
# This notebook reads a vertical profile data file and plots various atmospheric variables.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- File path ---
file_path = "output/arm_shcu_t_21600.dat"  # Change this if loading a different file


# --- Read file ---
with open(file_path, 'r') as f:
    lines = f.readlines()

# --- Parse global parameters from the first line ---
u_star, w_star, l_obuk, h_bl, th_surf, wth, wqv, zcl, z_cb, z_ct = map(float, lines[0].split())

print("Global Parameters:")
print(f"u_star = {u_star}")
print(f"w_star = {w_star}")
print(f"l_obuk = {l_obuk}")
print(f"h_bl = {h_bl}")
print(f"th_surf = {th_surf}")
print(f"w'theta' = {wth}")
print(f"w'q' = {wqv}")
print(f"zcl = {zcl}")
print(f"z_cb = {z_cb}")
print(f"z_ct = {z_ct}")


# --- Column names ---
columns = ['zn', 'z', 'p', 'u', 'v', 'th', 'qv', 'ql', 'rh', 'uw', 'vw', 
           'wth', 'k_m', 'k_h', 'c_vis']

# --- Load profile data ---
# --- Filter out non-numeric lines (e.g., footer text like '*********Next time ****************') ---
data_lines = [line for line in lines[1:] if line.strip() and not line.lstrip().startswith('*')]

# --- Load only valid numeric data, using the first 12 columns ---
data = np.loadtxt(data_lines, usecols=range(15))
df = pd.DataFrame(data, columns=columns)


# --- Plotting ---
def plot_profile(x, y, xlabel, ylabel='z (m)', title=None):
    plt.figure()
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if title:
        plt.title(title)
    plt.grid(True)
    plt.ylim(0,2000)
    plt.show()

# Example plots
plot_profile(df['p'], df['z'], 'Pressure (Pa)', title='Pressure (Pa)')
plot_profile(df['u'], df['zn'], 'u (m/s)', title='Horizontal Wind (u)')
plot_profile(df['v'], df['zn'], 'v (m/s)', title='Horizontal Wind (v)')
plot_profile(df['th'], df['z'], 'Potential Temperature (K)', title='Potential Temperature')
plot_profile(df['qv'], df['z'], 'q_v (g/kg)', title='Vapour mixing ratio')
plot_profile(df['ql'], df['z'], 'q_l (g/kg)', title='Liquid mixing ratio')
plot_profile(df['rh'], df['z'], 'rh ', title='Liquid mixing ratio')
plot_profile(df['uw'], df['zn'], 'uw (m/s)', title='Momentum Flux uw')
plot_profile(df['vw'], df['zn'], 'vw (m/s)', title='Momentum Flux vw')
plot_profile(df['wth'], df['z'], 'w (Km/s)', title='Buoyancy Flux w')
plot_profile(df['k_m'], df['z'], 'k_m (m/s)', title='Eddy Viscosity k_m')
plot_profile(df['k_h'], df['z'], 'k_h (m/s)', title='Thermal Diffusivity k_h')
plot_profile(df['c_vis'], df['z'], 'c_vis', title='Turbulent Coefficient c_vis')

# Optional: Save to CSV
# df.to_csv("profile_data.csv", index=False)
```

```python
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Directory containing the files ---
directory = "output/"  # Change to the directory where your files are stored

# --- Regex pattern to match files ---
file_pattern = re.compile(r"arm_shcu_t_(\d+)\.dat")

# --- Find all matching files and extract time steps ---
file_info = []
for fname in os.listdir(directory):
    match = file_pattern.match(fname)
    if match:
        time = int(match.group(1))
        path = os.path.join(directory, fname)
        file_info.append((time, path))

# --- Sort files by time ---
file_info.sort()

# --- Initialize list to hold global parameter data ---
data = []

# --- Loop over all files and extract global parameters from first line ---
for time, path in file_info:
    with open(path, 'r') as f:
        first_line = f.readline().strip()
    try:
        u_star, w_star, l_obuk, h_bl, th_surf, wth, wqv, zcl, z_b, z_t = map(float, first_line.split())
        data.append([time, u_star, w_star, l_obuk, h_bl, th_surf, wth, wqv, zcl, z_b, z_t])
    except ValueError:
        print(f"Skipping file {path}: could not parse global parameters.")

# --- Convert to DataFrame ---
df = pd.DataFrame(data, columns=["time", "u_star", "w_star", "l_obuk", "h_bl", 
                                 "th_surf", "wth", "wqv", "zcl","z_b","z_t"])
df = df.sort_values("time")

# --- Plotting function ---
def plot_variable(var, ylabel):
    plt.figure()
    plt.plot(df["time"], df[var], marker='o')
    plt.title(f"Time Evolution of {var}")
    plt.xlabel("Time (s)")
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

# --- Plot all global variables ---
plot_variable("u_star", "u* (m/s)")
plot_variable("w_star", "w* (m/s)")
plot_variable("l_obuk", "L (Obukhov length) (m)")
plot_variable("h_bl", "Boundary Layer Height (m)")
plot_variable("th_surf", "Surface Potential Temperature (K)")
plot_variable("wth", "Surface Potential Temperature Heat Flux (K m/s)")
plot_variable("wqv", "Surface Latent Heat Flux (g/g m/s)")
plot_variable("zcl", "LCL (m)")
```
